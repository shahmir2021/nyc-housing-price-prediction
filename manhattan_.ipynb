```python
#. Drop unnecessary columns
usa_inflation = usa_inflation.drop(
    columns=['Series Name', 'Series Code', 'Country Name', 'Country Code'],
    errors='ignore')
```

```python
# . Extract year columns (e.g. '2013 [YR2013]', ..., '2024 [YR2024]')
year_cols = [c for c in usa_inflation.columns if 'YR' in c]
```

```python
# . Convert inflation rates to numeric
rates = usa_inflation.loc[0, year_cols].astype(float)
```

```python
rates
```

```python
# . Create inflation DataFrame
inflation_df = pd.DataFrame({
    'YEAR': [int(c.split()[0]) for c in year_cols],
    'INFLATION': rates.values  # <-- inflation rate for that year (%)
})
```

```python
inflation_df
```

```python
# . Compute cumulative inflation up to 2024
inflation_df = inflation_df.sort_values('YEAR').reset_index(drop=True)
inflation_df['rate_frac'] = inflation_df['INFLATION'] / 100
inflation_df['price_index'] = (1 + inflation_df['rate_frac']).cumprod()
```

```python
# . Calculate factor to adjust each year’s prices to 2024 values
price_index_2024 = inflation_df.loc[inflation_df['YEAR'] == 2024, 'price_index'].iloc[0]
inflation_df['inflation_factor'] = price_index_2024 / inflation_df['price_index']
```

```python
print("Inflation DataFrame:")
print(inflation_df[['YEAR', 'INFLATION', 'inflation_factor']])
```

```python
pd.set_option('display.float_format', '{:,.0f}'.format)
```

```python
#  grab the year columns from usa_inflation, they look like "2013 [YR2013]" etc
year_cols = [c for c in usa_inflation.columns if pd.Series(c).str.contains(r'^\s*\d{4}\s*\[YR\d{4}\]', regex=True).item()]
```

```python
# rates row, assumes your file has one row of rates
rates_row = usa_inflation.loc[0, year_cols].astype(float)
```

```python
# extract the year number, make a tidy frame
years = pd.Series(year_cols).str.extract(r'(\d{4})', expand=False).astype(int).to_list()
infl = pd.DataFrame({'YEAR': years, 'rate_pct': rates_row.values})
```

```python
#  build a cumulative price index, base at first year, then compute factor to 2024
infl = infl.sort_values('YEAR').reset_index(drop=True)
infl['rate_frac'] = infl['rate_pct'] / 100.0
infl['index'] = (1.0 + infl['rate_frac']).cumprod()
```

```python
# pick 2024 as target year, adjust if your target is different
target_year = 2024
target_idx = infl.loc[infl['YEAR'] == target_year, 'index'].iloc[0]
infl['to_2024_factor'] = target_idx / infl['index']
```

### Deep and Cross Network

```python
from tensorflow.keras.layers import Layer
```

```python
class CrossLayer(Layer):
    def __init__(self, num_layers, **kwargs):
        super().__init__(**kwargs)
        self.num_layers = num_layers
        self.w = []
        self.b = []

    def build(self, input_shape):
        input_dim = int(input_shape[-1])
        for i in range(self.num_layers):
            self.w.append(
                self.add_weight(
                    shape=(input_dim, 1),
                    initializer="glorot_uniform",
                    name=f"w_{i}"
                )
            )
            self.b.append(
                self.add_weight(
                    shape=(input_dim,),
                    initializer="zeros",
                    name=f"b_{i}"
                )
            )

    def call(self, x0):
        x = x0
        for i in range(self.num_layers):
            # xw shape: (batch, 1)
            xw = tf.matmul(x, self.w[i])   # linear projection
            # broadcast to (batch, input_dim) and elementwise multiply with x0
            cross = x0 * xw + self.b[i]
            x = x + cross
        return x


def build_dcn(input_dim, cross_layers=3, deep_units=[128, 64, 32]):
    inp = Input(shape=(input_dim,))

    # cross network
    cross_out = CrossLayer(num_layers=cross_layers, name="cross_layer")(inp)

    # deep network
    deep = inp
    for i, units in enumerate(deep_units):
        deep = Dense(units, activation="relu", name=f"deep_dense_{i}")(deep)

    # use Keras Concatenate instead of tf.concat
    combined = Concatenate(name="combine_cross_deep")([cross_out, deep])

    out = Dense(1, name="output")(combined)

    model = Model(inputs=inp, outputs=out, name="DeepAndCrossNetwork")
    model.compile(optimizer="adam", loss="mse", metrics=["mae"])
    return model
```

### RIDGE

```python

class RidgeScratch:
    def __init__(self, alpha=1.0):
        """
        Simple Ridge Regression from scratch, with:
        - L2 penalty on weights
        - no penalty on intercept
        - internal standardisation of X
        """
        self.alpha = float(alpha)
        self.coef_ = None           # shape (n_features,)
        self.intercept_ = None      # scalar
        self.x_mean_ = None
        self.x_std_ = None
        self.is_fitted_ = False

    def _standardise(self, X):
        """
        Standardise features using training mean and std.
        Avoid division by zero if a feature has zero variance.
        """
        X = np.asarray(X, dtype=float)
        X_std_safe = np.where(self.x_std_ == 0, 1.0, self.x_std_)
        return (X - self.x_mean_) / X_std_safe

    def fit(self, X, y):
        """
        Closed form ridge:
        beta = (X^T X + alpha I)^-1 X^T y
        with:
        - design matrix [1, X_standardised]
        - intercept column not regularised
        """
        X = np.asarray(X, dtype=float)
        y = np.asarray(y, dtype=float).reshape(-1, 1)

        n_samples, n_features = X.shape

        # compute mean and std on training data
        self.x_mean_ = X.mean(axis=0)
        self.x_std_ = X.std(axis=0)

        # standardise X
        Xs = self._standardise(X)                      # shape (n_samples, n_features)

        # design matrix with bias column
        X_design = np.concatenate(
            [np.ones((n_samples, 1)), Xs],
            axis=1
        )                                              # shape (n_samples, n_features + 1)

        # ridge matrix, do not regularise intercept term
        n_params = X_design.shape[1]
        I = np.eye(n_params)
        I[0, 0] = 0.0                                  # no penalty on intercept

        # closed form ridge solution
        A = X_design.T @ X_design + self.alpha * I
        b = X_design.T @ y
        beta = np.linalg.solve(A, b)                   # shape (n_features + 1, 1)

        # store parameters
        self.intercept_ = float(beta[0, 0])
        self.coef_ = beta[1:, 0]                       # shape (n_features,)

        self.is_fitted_ = True
        return self

    def predict(self, X):
        """
        Predict using the fitted ridge model.
        """
        if not self.is_fitted_:
            raise RuntimeError("You must call fit before predict.")

        X = np.asarray(X, dtype=float)
        Xs = self._standardise(X)

        # y = intercept + X_standardised · coef
        return self.intercept_ + Xs @ self.coef_
```

```python
y_pred_train = ridge_scratch.predict(X_train)
y_pred_test  = ridge_scratch.predict(X_test)

train_mae = mean_absolute_error(y_train, y_pred_train)
test_mae  = mean_absolute_error(y_test, y_pred_test)

train_rmse = np.sqrt(mean_squared_error(y_train, y_pred_train))
test_rmse  = np.sqrt(mean_squared_error(y_test, y_pred_test))

train_r2 = r2_score(y_train, y_pred_train)
test_r2  = r2_score(y_test, y_pred_test)

print("RidgeScratch (alpha=100)")
print("Train MAE :", train_mae)
print("Test  MAE :", test_mae)
print("Train RMSE:", train_rmse)
print("Test  RMSE:", test_rmse)
print("Train R2  :", train_r2)
print("Test  R2  :", test_r2)
```
